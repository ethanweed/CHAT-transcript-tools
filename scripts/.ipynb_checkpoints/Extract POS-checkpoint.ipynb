```python
#read file into python

import os
pathin = '/Users/ethan/Dropbox/Research/Projects/Projects_current/uconn/text_processing/raw_cohorts_2_3/'
file = 'AD.Visit1.txt'


os.chdir(pathin)

removelist = ['\t', '\r']

```

```python
with open(file,'r') as f:
    text = f.read()
    for item in removelist:
        text = text.replace(item, '')
    text = text.split('@')
    
    #print(text)
    
    for line in text:
        if line.startswith('Participants:'):
            p = line.replace('Participants:', '')
            p = p.split(',')
            
            p1 = []
            for i, val in enumerate(p):
                v1 = val.replace(' ','')
                v2 = v1.replace('\n', '')
                p1.append(v2[0:3])
    people = p1
```

```python
with open(file,'r') as f:
    text = f.read()
    for item in removelist:
        text = text.replace(item, '')
    text = text.split('\n')

    trans = []
    for line in text:
        label = line[1:4]
        if label in people:
            trans.append(line)
        elif label == 'mor':
            trans.append(line)
print(trans)
```

```python
grouped = [trans[n:n+2] for n in range(0, len(trans), 2)]
grouped
```

```python
turn = []
speaker = []
for i, val in grouped:
    turn.append(i)

for i in grouped:
    s = i[0]
    m = i[1]
    speaker.append(s[1:4])
```
